```python
class FirstDay:

    def __init__(self, series):
        self.series = series
        self.len = len(series)

    def mean(self):

        value = (self.len / sum(self.series))*100

        return round(value, 2)

    def trimmed_mean(self, trim_percent=20):

        lst = self.series
        
        if not lst:
            raise ValueError("List is empty")
    
        if not (0 <= trim_percent < 50):
            raise ValueError("Invalid trim percentage")
    
        num_elements_to_trim = int(len(lst) * trim_percent / 100)
        sorted_lst = sorted(lst)
        trimmed_lst = sorted_lst[num_elements_to_trim : -num_elements_to_trim]
        trimmed_mean_value = sum(trimmed_lst) / len(trimmed_lst)
    
        return trimmed_mean_value

    def median(self):

        mid = self.len // 2

        self.series = sorted(self.series)
        
        for i in range(self.len):
            if self.len % 2 == 0:
                value = (self.series[mid] + self.series[mid + 1]) / 2
            elif self.len % 2 != 0:
                value = self.series[mid]

        return value

    def mode(self):

        vals = {}

        for i in range(self.len):
            # Initialize count to 1 for each unique element
            if self.series[i] not in vals:
                vals[self.series[i]] = 1
            else:
                vals[self.series[i]] += 1
    
        # Find the key with the maximum count
        max_count_key = max(vals, key=vals.get)

        return max_count_key

    def range(self):

        range = max(self.series) - min(self.series)

        return range

    def variance(self):

        sigma_sq = 0

        for i in range(self.len):
            sigma_sq += (self.series[i] - self.mean()) ** 2

        sigma_sq = sigma_sq / self.len

        return round(sigma_sq, 2)

    def stand_dev(self):

        std = (self.variance()) ** 0.5

        return round(std, 2)

    def coef_var(self):

        coef = (self.stand_dev() / self.trimmed_mean()) * 100

        return coef
```
