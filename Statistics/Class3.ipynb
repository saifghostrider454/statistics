```python
import pandas as pd
import numpy as np
import seaborn as sns
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class ProbabilityMassFunction:

    def oneDiceRoll(self, number_of_rolls, plot=False, cumm=False):
        """PMF FUNCTION"""
        L = []
        for i in range(number_of_rolls):
            L.append(random.randint(1, 6))

        series = pd.Series(L)
        probability = (series.value_counts() / series.value_counts().sum()).sort_index()

        if plot == True:
            plt.figure(figsize=(4, 2))
            num_bars = len(probability)
            colors = sns.color_palette('husl', n_colors=num_bars)
            probability.plot(kind='bar', color=colors)
            plt.xticks(rotation=360)
            plt.xlabel('Possible Outcome')
            plt.ylabel('Probability')
            plt.title('Probability Mass Function (PMF)')

        if cumm == True:
            plt.figure(figsize=(4, 2))
            num_bars = len(probability)
            colors = sns.color_palette('husl', n_colors=num_bars)
            np.cumsum(probability).plot(kind='bar', color=colors)
            plt.xticks(rotation=360)
            plt.xlabel('Possible Outcome')
            plt.ylabel('Probability')
            plt.title('Cumulative Distribution Function (CDF of PMF)')
        return probability

    def twoDiceRoll(self, number_of_rolls, plot=False, cumm=False):
        """PMF FUNCTION"""
        L = []
        for i in range(number_of_rolls):
            a = random.randint(1, 6)
            b = random.randint(1, 6)

            L.append(a + b)

        series = pd.Series(L)
        probability = (series.value_counts() / series.value_counts().sum()).sort_index()

        if plot == True:
            plt.figure(figsize=(4, 2))
            num_bars = len(probability)
            colors = sns.color_palette('husl', n_colors=num_bars)
            probability.plot(kind='bar', color=colors)
            plt.xticks(rotation=360)
            plt.xlabel('Possible Outcome')
            plt.ylabel('Probability')
            plt.title('Probability Mass Function (PMF)')

        if cumm == True:
            plt.figure(figsize=(4, 2))
            num_bars = len(probability)
            colors = sns.color_palette('husl', n_colors=num_bars)
            np.cumsum(probability).plot(kind='bar', color=colors)
            plt.xticks(rotation=360)
            plt.xlabel('Possible Outcome')
            plt.ylabel('Probability')
            plt.title('Cumulative Distribution Function (CDF of PMF)')
        return probability
```

```python
test = ProbabilityMassFunction()
```

```python
test.oneDiceRoll(3000, plot=True, cumm=True)
```

```python
test.twoDiceRoll(3000, plot=True, cumm=True)
```
